# EXPLANATORY DATA ANALYSIS (EDA) #

## ANALYSIS OF TOP PRODUCTS ##

```python
df['Product'].value_counts()
```

```python
top_products=df['Product'].value_counts().nlargest(5)
df_top_products=df[df['Product'].isin(top_products.index)]
```

```python
plt.figure(figsize=(13,5))
sns.countplot(data=df_top_products, x='Product', order=top_products.index, palette='viridis',legend=False)
plt.title("Top 5 Products by Count")
plt.ylabel("Count")
plt.xlabel("Product")
plt.show()
```

* As we can see, the top selling product is 2GB Graphic Card, which has been sold almost twice than the other leading products.

## ANALYSIS CO-RELATING CATEGORY AND BRANDS ##

```python
top_cat=df['Category'].value_counts().nlargest(10)
top_bran=df['Brand'].value_counts().nlargest(5)
df_top_catbran=df[df['Category'].isin(top_cat.index)]
df_top_catbran=df[df['Brand'].isin(top_bran.index)]
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Category', data=df_top_catbran, hue='Brand', palette='viridis')
plt.show()
```

* A very interesting data comes up here. We can see that consumers are mostly prefering a certain brand for a certain category. It is clear from the visuals that Samsung is the top choice when it comes to SSD and Mouse. Intel is the only one dominating the CPU market whereas Nvidia is the major player when it comes to Graphic Card. There is competition between Samsung and Hynix in the RAM Category.Samsung is also present in the monitor market but Dell is the frontrunner there. Hynix has its in Computer Case and NIC. This shows a very practical scenario. 

## ANALYSIS CO-RELATING TOTAL SALES AND CATEGORY ##

```python
plt.figure(figsize=(12,6))
top_cat=df['Category'].value_counts().nlargest(10)
df_top_catsal=df[df['Category'].isin(top_cat.index)]
sales=df_top_catsal.groupby(['Category'], as_index=False)['Total_Sales'].sum().sort_values(by='Total_Sales', ascending=False)
sns.barplot(x='Category', y='Total_Sales', data=sales, palette='Blues')
plt.show()
```

* Here it is clearly visible that Monitors bring the most revenue followed by CPU. It must be noted that Monitors and CPU also topped the chart of number of orders. However, Graphic Cards which bring the third largest revenue are fifth in terms of number of orders. Thus, if more Graphics Card can be sold then it will significantly boost the revenue due to its high order value. The opposite scenario can be seen with Mouse which is the third best selling item in terms of number of orders but brings much less revenue due to low price.

## ANALYSIS OF YEAR ON YEAR GROWTH ##

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'], format = '%d/%m/%Y')
```

```python
df['Order Year']=df['Order_Date'].dt.year
```

```python
df['Order Year']
```

```python
df['Order Year'].value_counts()
```

```python
x=df['Order Year'].value_counts().index
y=df['Order Year'].value_counts().values
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,'o:r')
plt.title("YoY Order Growth")
plt.show()
```

* It is visible that the number of orders have increased every year
